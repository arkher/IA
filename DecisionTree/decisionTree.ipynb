```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_excel('Data/dataset_jogadores.xlsx')
```

```python
df_para_responder = df.loc[df.Classe == '?']
df_para_aprendizado = df.loc[df.Classe != '?']
df_para_aprendizado = df_para_aprendizado.loc[df_para_aprendizado.Classe != 'Classe']
df_para_aprendizado = df_para_aprendizado.dropna(thresh=2)

df_para_aprendizado['classe_suporte'] = (df_para_aprendizado.Classe == 'Defensor').astype('int')
df_para_aprendizado = df_para_aprendizado.drop(columns=['Classe'])
```

```python
df_para_aprendizado.classe_suporte.value_counts()
```

```python
clf =tree.DecisionTreeClassifier()
```

```python
df_para_responder.head()
def para_nome(list):
    retorno = []
    for element in list:
        if(element == 1):
            retorno.append('Defensor')
        else:
            retorno.append('Atacante')
    return retorno
```

```python
def especificidade(list1,list2):
    count =0
    for posi in range(len(list1)):
        if(list1[posi] == 0 and  list2[posi] == list1[posi]):
            count = count+1
    return count/(len(list1)-sum(list1))
def sensibilidade(list1,list2):
    count =0
    for posi in range(len(list1)):
        if(list1[posi] == 1 and  list2[posi] == list1[posi]):
            count = count+1
    return count/(sum(list1))
```

```python
target = df_para_aprendizado.classe_suporte
treino_x,teste_x,treino_y,teste_y = train_test_split(df_para_aprendizado.drop(columns=['classe_suporte','Id']), target)
    
clf = clf.fit(treino_x,treino_y)
prediction = clf.predict(teste_x.values)
score = accuracy_score(teste_y, prediction)

from sklearn.metrics import confusion_matrix

cm1 = confusion_matrix(teste_y , prediction)
print('Confusion Matrix : \n', cm1)

total1=sum(sum(cm1))
accuracy1=(cm1[0,0]+cm1[1,1])/total1
print ('Accuracy : ', accuracy1)

sensitivity1 = cm1[0,0]/(cm1[0,0]+cm1[0,1])
print('Sensitivity : ', sensitivity1 )

specificity1 = cm1[1,1]/(cm1[1,0]+cm1[1,1])
print('Specificity : ', specificity1)

```

```python
df_para_responder = df_para_responder.drop(columns = ['Classe', 'classe_suporte'],errors='ignore')
predictions = clf.predict(df_para_responder.drop(columns=['Id']).values)
df_para_responder['Classe'] = para_nome(predictions)
```

```python
df_para_responder.to_csv('saida.csv')
df_para_responder
```
